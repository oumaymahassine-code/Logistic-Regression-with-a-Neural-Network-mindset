```python
import kagglehub
path = kagglehub.dataset_download("tuanai/traffic-signs-dataset")
```

```python
import tensorflow as tf
from tensorflow.keras import layers, models
import matplotlib.pyplot as plt
```

```python
import os

# Charger le dataset depuis les dossiers

IMG_SIZE = 32
BATCH_SIZE = 32

# Assuming 'path' variable from previous cell contains the base path to the downloaded dataset
# Temporarily list contents to debug directory structure
print(f"Contents of {path}:")
for item in os.listdir(path):
    print(f" - {item}")

# The previous data_directory caused a FileNotFoundError.
# Based on the output of os.listdir(path), 'DATA' is a direct subdirectory of 'path'.
# So, the correct path is os.path.join(path, 'DATA').
data_directory = os.path.join(path, "DATA")

print(f"Attempting to load data from: {data_directory}")

train_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    data_directory,
    validation_split=0.2,
    subset="training",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)

test_dataset = tf.keras.preprocessing.image_dataset_from_directory(
    data_directory,
    validation_split=0.2,
    subset="validation",
    seed=123,
    image_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE
)
```

```python
# Nombre de classes
class_names = train_dataset.class_names
num_classes = len(class_names)

print("Classes :", class_names)
print("Nombre de classes :", num_classes)
```

```python
# Normalisation
normalization_layer = layers.Rescaling(1./255)

train_dataset = train_dataset.map(lambda x, y: (normalization_layer(x), y))
test_dataset = test_dataset.map(lambda x, y: (normalization_layer(x), y))
```

```python
# Modèle CNN
model = models.Sequential([

    layers.Conv2D(
        7,
        kernel_size=(3,3),
        activation='relu',
        input_shape=(32,32,3)
    ),

    layers.MaxPooling2D((2,2)),

    layers.Flatten(),

    layers.Dense(225, activation='relu'),

    layers.Dense(num_classes, activation='softmax')
])
```

```python
# Compiler le modèle
model.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
```

```python
# Entraînement
history = model.fit(
    train_dataset,
    epochs=5,
    validation_data=test_dataset
)
```

```python
# Évaluation
loss, accuracy = model.evaluate(test_dataset)

print("Accuracy :", accuracy)
```
